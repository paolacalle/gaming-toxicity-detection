### Consolidated Experiment Results

The tables below rank each experiment by `CV Macro F1`, then `CV Weighted F1`, then `Accuracy`. The first table uses validation metrics; the second table uses train metrics.


```python
# Consolidated experiment results
metric_order = [
    "CV Macro F1",
    "CV Weighted F1",
    "Accuracy",
    "Coverage",
    "Precision",
    "FPR",
    "FNR",
    "TPR",
    "TNR",
]


def build_results_table(split_name):
    rows = []
    for item in experiments:
        metrics = item[split_name]
        row = {
            "Experiment": item["experiment"],
            "Models": ", ".join(item["models"]),
        }
        for metric in metric_order:
            if metric in metrics:
                row[metric] = float(metrics[metric])
        rows.append(row)

    table = pd.DataFrame(rows)
    metric_cols = [metric for metric in metric_order if metric in table.columns]
    table = table.sort_values(
        by=["CV Macro F1", "CV Weighted F1", "Accuracy"],
        ascending=False,
        na_position="last",
    ).reset_index(drop=True)
    table.insert(0, "Rank", range(1, len(table) + 1))
    return table, metric_cols


validation_results_table, validation_metric_cols = build_results_table("validation")
train_results_table, train_metric_cols = build_results_table("train")

validation_results_table[["Rank", "Experiment", "Models", *validation_metric_cols]].style.format(
    {metric: "{:.4f}" for metric in validation_metric_cols}, na_rep=""
)
```
